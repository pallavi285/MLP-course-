# Preprocessing

    **encoding**

Why Encoding is Necessary
Most machine learning models, especially algorithms like linear regression, logistic regression, and neural networks, cannot handle raw categorical or text data directly. They require numerical input. Encoding transforms non-numeric data into a format that models can understand while preserving the information inherent in the variables.

# categorical data
- Nominal data    
       Nominal data is a type of qualitative data used to represent distinct categories or groups without any order. It is often analyzed using frequencies, percentages, or mode, and the categories do not overlap or have a specific ranking. Examples of nominal data include gender, ethnicity, or color, where the data is classified into distinct categories that cannot be ordered or measured.
